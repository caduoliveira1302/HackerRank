```python
def countApplesAndOranges(s:int, t:int, a:int, b:int, apples:list, oranges: list):
    #renaming variables for better understanding 
    house_range = [s,t]
    apple_tree_loc = a
    orange_tree_loc = b
    #calcuting distance for each fruit from trees
    distance_AppleTree_House = [apple_tree_loc+apples[i] for i in range(len(apples))]
    distance_OrangeTree_House = [orange_tree_loc+oranges[i] for i in range(len(oranges))]

    #How many oranges&apples land in house range?
    apples_in_house_range = 0
    oranges_in_house_range = 0

    for apple in distance_AppleTree_House:
        if  house_range[1] >= apple >= house_range[0]:
            apples_in_house_range += 1
    for orange in distance_OrangeTree_House:
        if house_range[1] >= orange >= house_range[0]:
            oranges_in_house_range += 1

    return print(f'{apples_in_house_range}\n{oranges_in_house_range}')
```

```python
countApplesAndOranges(7, 10, 4, 12, [2,3,-4], [3,-2,-4])
```
