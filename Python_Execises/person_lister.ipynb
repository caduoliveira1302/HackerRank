```python
import pandas as pd
from operator import itemgetter
```

##### Working with dataframes (1st ideia I had)

```python
def person_lister(people_list):
    df = pd.DataFrame({'Name' : [people_list[i][0] for i in range(len(people_list))],
                    'Age' : [people_list[i][1] for i in range(len(people_list))],
                    'Sex' : [people_list[i][2] for i in range(len(people_list))]})
    df = df.sort_values(by='Age')

    for value in df.values:
        if value[2] == 'F':
            print(f'Ms. {value[0]}')
        else:
            print(f'Mr. {value[0]}')
```

##### Working without dataframes, using *operator* library 

```python
def person_lister2(people_list):
    people_list.sort(key=itemgetter(1)) #sorting by index 1 (people_list[0][1] --> age)

    for p in people_list:
        if p[2] == 'F':
            print(f'Ms. {p[0]}')
        else:
            print(f'Mr. {p[0]}')
```

##### RESULTS

```python
people_list = [
    ['Mark Brown', 52, 'M'], 
    ['Mary Cooper', 32, 'F'], 
    ['Bruno Mars', 40, 'M'], 
    ['Martha Sprite', 32, 'F'], 
    ['Sayd Oliver', 60, 'M']
    ]
```

```python
person_lister(people_list)
```

```python
person_lister2(people_list)
```
