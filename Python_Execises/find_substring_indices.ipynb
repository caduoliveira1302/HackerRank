```python
def find_substring_indices(S, k):
    start = 0
    found = False  
    
    while True:
        start = S.find(k, start)
        if start == -1:
            break
        end = start + len(k) - 1
        print(f"({start}, {end})")
        found = True
        start += 1
    if not found:
        print("(-1, -1)")
```

```python
S = 'aaadaa'
k = 'aa'
find_substring_indices(S, k)
```

```python
S = 'aaadaa'
k = 'b'
find_substring_indices(S, k)
```
