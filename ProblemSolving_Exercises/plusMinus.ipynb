```python
def plusMinus(arr):
    less_0 = 0
    more_0 = 0
    equal_0 = 0

    for value in arr:
        if value < 0:
            less_0 += 1
        elif value > 0:
            more_0 += 1
        elif value == 0:
            equal_0 += 1

    return print(f" {more_0/len(arr):.6f} \n {less_0/len(arr):.6f} \n {equal_0/len(arr):.6f}")
```

```python
arr = [2,9,10,0,-3,-4,0,2,-9,1]
plusMinus(arr)
```
