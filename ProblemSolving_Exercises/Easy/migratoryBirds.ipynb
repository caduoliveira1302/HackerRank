```python
def migratoryBirds(arr):
    unique_id_list = list(set(arr)) #unique ID's
    quantity_list = []              #number of time the ID appears

    for id in unique_id_list:
        quantity_list.append(arr.count(id))

    #number of times that the most appear id appears
    max_qtt = max(quantity_list)
    #the indexes of the most appear id
    #if there is more than 1 index, we will pick the lower one.
    indexes = [i for i, x in enumerate(quantity_list) if x == max_qtt]

    return unique_id_list[indexes[0]] #always picking the lower id (indexes[0])
```

```python
migratoryBirds([1,2,3,4,5,4,3,2,1,3,4])
```
