```python
def icecreamParlor(m: int, arr: list):
    for n1 in range(0, len(arr), 1):
        for n2 in range(1, len(arr), 1):
            if arr[n1]+arr[n2] == m:
                return print(n1+1, n2+1)
        break #interrupt the second output, I only need the first output
```

```python
icecreamParlor(4, [2,2,4,3])
```
