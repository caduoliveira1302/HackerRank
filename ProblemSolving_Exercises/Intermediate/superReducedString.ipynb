```python
def superReducedString(s: str):
    letters = list(set(s))
    remove = [letter*2 for letter in letters]
        
    for _ in range(len(s)):
        for r in remove:
            if r in s:
                s = s.replace(r, '')
            else:
                pass
    if len(s) == 0:
        return 'Empty String'
    else:
        return s
```

```python
superReducedString('aaabccddd')
```
