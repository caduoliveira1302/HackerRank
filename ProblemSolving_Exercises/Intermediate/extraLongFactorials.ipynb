```python
def extraLongFactorials(n):
    factorial = n
    before_num = n - 1
    result = factorial * before_num

    if n == 1:
        return print(1)
    else:
        for i in range(2, n, 1):
            while before_num != 1:
                result = result * (before_num - 1)
                before_num = before_num - 1
        
        return print(result)
```

```python
extraLongFactorials(25)
```
