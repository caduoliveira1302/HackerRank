```python
def pairs(k:int, arr:list):
    count_pairs = 0

    for number in arr:
        resultado = [number - i for i in arr]
        count_pairs += resultado.count(k)

    return count_pairs
```

```python
pairs(2, [1,5,3,4,2])
```

```python
pairs(1, [1,2,3,4])
```

```python
pairs(2, [1,3,5,8,6,4,2])
```
