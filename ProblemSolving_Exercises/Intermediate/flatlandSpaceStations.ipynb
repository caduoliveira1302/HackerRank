```python
def flatlandSpaceStations(n: int, c: int):
    cities = [city for city in range(n)]
    space_st = c

    #in case of every city has a space station
    if n == len(c):
        return 0
    else:
        distances = []
        for city in cities:
            #verifying the nearst space station for each city
            diffs = [abs(city - ss) for ss in space_st]
            nearst_ss = space_st[diffs.index(min(diffs))]
            #calculating the distance from city to the nearst space station, appending it to the distances list and returning the max distance
            distance = abs(city - nearst_ss)
            distances.append(distance)

        print(max(distances))
```

```python
flatlandSpaceStations(5, [0,4])
```

```python
flatlandSpaceStations(6, [0,1,2,3,4,5])
```

```python
flatlandSpaceStations(99998, [28000, 58701, 43043, 24909, 28572])
```
