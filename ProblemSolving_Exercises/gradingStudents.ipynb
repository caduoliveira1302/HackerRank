```python
import math
```

```python
def gradingStudents(grades):
    threshold = 3

    for grade in grades:
        if grade % 5 == 0: #grades divided by 5 dont need to be changed
            pass
        elif grade < 38: #grades less than 38 dont need to be changed
            pass 
        elif grade % 5 != 0:
            new_grade = math.ceil(grade/5)*5 #math.ceil() brings the nearest integer number
            sub_result = new_grade - grade 
            if sub_result < threshold:
                grades[grades.index(grade)] = new_grade
            else:
                pass

    
    for grade in grades: print(grade)
```

```python
arr = [73, 67, 38, 33]
gradingStudents(arr)
```
