```python
def kangaroo(x1: int, v1: int, x2: int, v2:int):
    #Uptading x1 and x2 values by their jumps values. If x1 and x2 got the same value, we print YES
    for i in range(10000):
        x1+=v1
        x2+=v2
        if x1==x2:
            return 'YES'
    #After 10000 tentatives, if x1 and x2 are still different, we print NO
    if x1!=x2:
        return 'NO'
```

```python
def kangaroo_fs(x1: int, v1: int, x2: int, v2: int):
    #if one kangaroo is in a bigger position and jumps longer, the other one can't get him.
    if (x1 < x2 and v1 <= v2) or (x2 < x1 and v2 <= v1): 
        return "NO"
    #The kangaroos will only meet if their difference in positions is divisible by the difference in their velocities.
    if (v1 != v2) and ((x2 - x1) % (v1 - v2)) == 0:
        return "YES"
    else:
        return "NO"
```

```python
display( kangaroo_fs(0,2,5,3) )
display( kangaroo_fs(2,1,1,2) )
```
