```python
def diagonalDifference(arr):
    main_diag_sum = 0
    sec_diag_sum = 0

    n = len(arr)

    for i in range(n):
        main_diag_sum += arr[i][i]
        sec_diag_sum += arr[i][n - i - 1]

    return abs(main_diag_sum-sec_diag_sum)
```

```python
arr = [
    [11, 2, 4],
    [4, 5, 6],
    [10, 8, -12]
]

diagonalDifference(arr)
```
