```python
class Item:
    def __init__(self, name: str, price: int):
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add(self, item: Item):
        self.items.append(item)

    def total(self) -> int:
        return sum(item.price for item in self.items)

    def __len__(self):
        return len(self.items)
```

```python
cart = ShoppingCart()
item1 = Item("Apple", 10)
item2 = Item("Banana", 5)

cart.add(item1)
cart.add(item2)

print(f"Total price: {cart.total()}")  
print(f"Number of items: {len(cart)}")  
```
